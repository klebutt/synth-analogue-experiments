# 2. Model Overview

Let's examine what models we have available and their configurations. This will help us understand the different forecasting approaches we can use.

```python
# Create model instances
models = {
    'Random Walk': RandomWalkModel(volatility=0.02),
    'Geometric Brownian Motion': GeometricBrownianModel(drift=0.001, volatility=0.02),
    'Mean Reversion': MeanReversionModel(mean_price=50000.0, reversion_strength=0.1, volatility=0.02)
}

# Display model information
print("üìä Available Models:")
print("=" * 50)
for name, model in models.items():
    info = model.get_model_info()
    print(f"\nÔøΩÔøΩ {name}")
    print(f"   Type: {info['type']}")
    print(f"   Description: {info['description']}")
    for key, value in info.items():
        if key not in ['name', 'type', 'description']:
            print(f"   {key}: {value}")
```

# 3. Generate Sample Predictions

Now let's generate predictions from all models to see how they work. We'll create realistic price forecasts that we can later evaluate using CRPS.

```python
# Set up prediction parameters
start_price = 50000.0  # Starting price (e.g., BTC)
start_time = datetime.now()
time_increment = 300   # 5 minutes in seconds
time_horizon = 3600    # 1 hour in seconds
num_simulations = 50   # Number of simulation paths

print(f"üéØ Prediction Parameters:")
print(f"   Start Price: ${start_price:,.2f}")
print(f"   Start Time: {start_time.strftime('%Y-%m-%d %H:%M:%S')}")
print(f"   Time Increment: {time_increment} seconds ({time_increment/60:.1f} minutes)")
print(f"   Time Horizon: {time_horizon} seconds ({time_horizon/3600:.1f} hours)")
print(f"   Simulations: {num_simulations}")
```

```python
# Generate predictions from all models
predictions = {}

for name, model in models.items():
    print(f"üîÑ Generating predictions for {name}...")
    
    try:
        model_predictions = model.predict(
            start_price=start_price,
            start_time=start_time,
            time_increment=time_increment,
            time_horizon=time_horizon,
            num_simulations=num_simulations
        )
        
        predictions[name] = model_predictions
        print(f"   ‚úÖ Generated {len(model_predictions)} simulations with {len(model_predictions[0])} time points")
        
    except Exception as e:
        print(f"   ‚ùå Error: {e}")

print(f"\nüéâ Successfully generated predictions for {len(predictions)} models!")
```

# 4. Visualize Price Predictions

Let's create visualizations to compare how different models predict price movements. This will help us understand the characteristics of each forecasting approach.

```python
# Helper function to extract time and price data
def extract_plot_data(predictions, model_name):
    """Extract time and price data for plotting."""
    times = []
    prices = []
    
    for sim_idx, simulation in enumerate(predictions):
        for point in simulation:
            time = datetime.fromisoformat(point['time'])
            price = point['price']
            
            times.append(time)
            prices.append(price)
    
    return times, prices

# Create visualization
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle('Price Predictions Comparison', fontsize=16, fontweight='bold')

for idx, (model_name, model_predictions) in enumerate(predictions.items()):
    row = idx // 2
    col = idx % 2
    ax = axes[row, col]
    
    # Plot individual simulation paths
    for sim_idx, simulation in enumerate(model_predictions):
        times = [datetime.fromisoformat(point['time']) for point in simulation]
        prices = [point['price'] for point in simulation]
        
        # Use different colors for different simulations
        alpha = 0.3 if sim_idx > 0 else 0.7
        ax.plot(times, prices, alpha=alpha, linewidth=1)
    
    # Plot mean path
    all_times = [datetime.fromisoformat(point['time']) for point in model_predictions[0]]
    mean_prices = []
    
    for time_idx in range(len(all_times)):
        time_prices = [sim[time_idx]['price'] for sim in model_predictions]
        mean_prices.append(np.mean(time_prices))
    
    ax.plot(all_times, mean_prices, 'k-', linewidth=3, label='Mean Path')
    
    ax.set_title(f'{model_name}', fontweight='bold')
    ax.set_xlabel('Time')
    ax.set_ylabel('Price ($)')
    ax.legend()
    ax.grid(True, alpha=0.3)
    
    # Format x-axis
    ax.tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()

print("üìà Visualization complete! Each subplot shows:")
print("   - Individual simulation paths (faded lines)")
print("   - Mean prediction path (bold black line)")
print("   - Different model behaviors and characteristics")
```

# 5. CRPS Calculation and Model Comparison

Now let's evaluate our models using CRPS (Continuous Ranked Probability Score), which is the primary metric used by Synth subnet to rank miners and allocate TAO rewards.

```python
# Create synthetic "actual" data for demonstration
# In real usage, this would come from actual market data
print("ÔøΩÔøΩ Creating synthetic actual data for CRPS evaluation...")

# Generate realistic actual prices based on our predictions
actual_data = []
base_time = start_time

for i in range(len(predictions[list(predictions.keys())[0]][0])):
    time_point = base_time + timedelta(seconds=i * time_increment)
    
    # Create realistic actual price with some market noise
    base_price = start_price + i * 50  # Slight upward trend
    market_noise = np.random.normal(0, 200)  # $200 standard deviation
    actual_price = max(base_price + market_noise, 1000.0)  # Ensure positive
    
    actual_data.append({
        'time': time_point.isoformat(),
        'price': actual_price
    })

print(f"‚úÖ Generated {len(actual_data)} actual price points")
print(f"   Time range: {actual_data[0]['time']} to {actual_data[-1]['time']}")
print(f"   Price range: ${min(p['price'] for p in actual_data):,.2f} to ${max(p['price'] for p in actual_data):,.2f}")
```

```python
# Calculate CRPS scores for all models
print("üìä Calculating CRPS scores...")
print("=" * 50)

crps_calculator = CRPSCalculator()
crps_results = {}

for model_name, model_predictions in predictions.items():
    try:
        metrics = crps_calculator.calculate_crps_for_synth(model_predictions, actual_data)
        crps_results[model_name] = metrics
        
        print(f"\nüîπ {model_name}")
        print(f"   CRPS Score: {metrics['crps_score']:.2f}")
        print(f"   Simulations: {metrics['num_simulations']}")
        print(f"   Time Points: {metrics['num_time_points']}")
        print(f"   Horizon: {metrics['prediction_horizon']} seconds")
        
    except Exception as e:
        print(f"\n‚ùå {model_name}: Error calculating CRPS - {e}")

print(f"\nüéØ CRPS calculation complete for {len(crps_results)} models!")
```

```python
# Compare models using CRPS
if crps_results:
    print("üèÜ Model Performance Ranking (Lower CRPS = Better):")
    print("=" * 60)
    
    # Sort models by CRPS score (lower is better)
    ranked_models = sorted(crps_results.items(), key=lambda x: x[1]['crps_score'])
    
    for rank, (model_name, metrics) in enumerate(ranked_models, 1):
        crps_score = metrics['crps_score']
        
        if rank == 1:
            medal = "ÔøΩÔøΩ"
        elif rank == 2:
            medal = "ÔøΩÔøΩ"
        elif rank == 3:
            medal = "ü•â"
        else:
            medal = "  "
        
        print(f"{medal} Rank {rank}: {model_name}")
        print(f"   CRPS Score: {crps_score:.2f}")
        print(f"   Performance: {'Excellent' if crps_score < 100 else 'Good' if crps_score < 500 else 'Fair'}")
        print()
    
    # Best model
    best_model = ranked_models[0][0]
    best_crps = ranked_models[0][1]['crps_score']
    
    print(f"üéØ Best Performing Model: {best_model}")
    print(f"   CRPS Score: {best_crps:.2f}")
    print(f"   This model would earn the most TAO rewards on Synth subnet!")
else:
    print("‚ùå No CRPS results available for comparison")
```

# 6. Interactive Parameter Exploration

Let's explore how different parameters affect model performance. This will help us optimize our models for maximum CRPS performance and TAO rewards on Synth subnet.

```python
# Test different volatility levels for Random Walk
print("üî¨ Testing Random Walk with different volatility levels...")

volatility_levels = [0.01, 0.02, 0.05, 0.10]
volatility_results = {}

for vol in volatility_levels:
    model = RandomWalkModel(volatility=vol)
    
    try:
        predictions = model.predict(
            start_price=start_price,
            start_time=start_time,
            time_increment=time_increment,
            time_horizon=time_horizon,
            num_simulations=20  # Fewer simulations for faster testing
        )
        
        metrics = crps_calculator.calculate_crps_for_synth(predictions, actual_data)
        volatility_results[vol] = metrics['crps_score']
        
        print(f"   Volatility {vol:.3f}: CRPS = {metrics['crps_score']:.2f}")
        
    except Exception as e:
        print(f"   Volatility {vol:.3f}: Error - {e}")

print(f"\n‚úÖ Tested {len(volatility_results)} volatility levels")
```

```python
# Visualize volatility impact
if volatility_results:
    plt.figure(figsize=(10, 6))
    
    volatilities = list(volatility_results.keys())
    crps_scores = list(volatility_results.values())
    
    plt.plot(volatilities, crps_scores, 'bo-', linewidth=2, markersize=8)
    plt.fill_between(volatilities, crps_scores, alpha=0.3)
    
    plt.title('Random Walk: Volatility vs CRPS Performance', fontweight='bold')
    plt.xlabel('Volatility')
    plt.ylabel('CRPS Score (Lower = Better)')
    plt.grid(True, alpha=0.3)
    
    # Add annotations
    for vol, crps in zip(volatilities, crps_scores):
        plt.annotate(f'{crps:.1f}', (vol, crps), 
                    textcoords="offset points", xytext=(0,10), ha='center')
    
    plt.tight_layout()
    plt.show()
    
    print("üìä Analysis:")
    best_vol = min(volatility_results.items(), key=lambda x: x[1])
    print(f"   Best volatility: {best_vol[0]:.3f} (CRPS: {best_vol[1]:.2f})")
    print(f"   Higher volatility generally increases CRPS (worse performance)")
    print(f"   Lower volatility provides more stable predictions")
```
